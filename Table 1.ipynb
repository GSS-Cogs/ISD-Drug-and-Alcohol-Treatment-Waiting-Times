Completed waits by age and gender table compiled from hidden tables Non Anon Totals

```python
from gssutils import *

if is_interactive():
    scraper = Scraper('http://www.isdscotland.org/Health-Topics/Drugs-and-Alcohol-Misuse/Publications/')
    scraper.select_dataset(title='National Drug and Alcohol Treatment Waiting Times')
    tabs = scraper.distribution(title='April 2014 - March 2018 Demographic Tables').as_databaker()
    tab = next(t for t in tabs if t.name == 'Non Anon Totals')
```

```python
cell = tab.filter('desc')
des = cell.expand(DOWN).is_not_blank()
Age = cell.shift(1,0).expand(RIGHT).is_not_blank()
observations = Age.expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
Dimensions = [
            HDim(des,'des',DIRECTLY,LEFT),
            HDim(Age,'Age',DIRECTLY,ABOVE),
            HDimConst('Measure Type', 'Count'),
            HDimConst('Unit','non-anon-clients'),
            HDimConst('Status', 'Number of non-anonymous clients who have started first treatment')
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
if is_interactive():
    savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table['OBS'] = new_table['OBS'].map(lambda x: '' if x == '*' else x)
```

```python
new_table.columns = ['Value' if x=='OBS' else x for x in new_table.columns]
```

```python
new_table['Age'] = new_table['Age'].map(lambda cell:cell.replace('Total', 'AGALL'))
```

```python
new_table['Trust'] = new_table['des'].astype(str).str[0:3]
```

```python
new_table['Client'] = new_table['des'].astype(str).str[3:4]
```

```python
new_table['Period'] = new_table['des'].astype(str).str[4:6]
```

```python
new_table['Sex'] = new_table['des'].astype(str).str[6:]
```

```python
new_table = new_table[new_table['Sex'] != '' ]
```

```python
new_table['Client'] = new_table['Client'].map(
    lambda x: {
        'a' : 'Alcohol', 
        'b' : 'Alcohol and Drugs',
        'd' : 'Drugs'
        }.get(x, x))
```

```python
new_table['Period'] = new_table['Period'].map(
    lambda x: {
        '11' : '31st March 2014', 
        '21' : '31st March 2015',
        '31' : '31st March 2016',
        '41' : '31st March 2017',
        '12' :  '31st March 2018'
        }.get(x, x))
```

```python
new_table = new_table[['Period','Age','Sex','Trust','Client','Status','Measure Type','Value','Unit']]
```
